```python
import pandas as pd #importar a biblioteca pandas
import seaborn as sns #importar a biblioteca para o gráfico seaborn
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay #matriz de confusão
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report #calcular accuracy, precision, recall, f1 score

iris = pd.read_csv("Iris.csv") #ler o arquivo

#MOSTRAR AS PRIMEIRAS 5 LINHAS DA TABELA
print(iris.head())
print("\n")

#MOSTRAR AS INFORMAÇÕES DA TABELA
iris.info()
print("\n")

#MOSTRAR AS CARACTERÍSTICAS DA TABELA
print(iris.describe())
print("\n")

#MOSTRAR APENAS A LINHA COM VALOR NULO
df_filtered_notna = iris[~iris['PetalWidthCm'].notna()]
print(df_filtered_notna)
print("\n")

#ADICIONAR O VALOR NO CAMPO NULO
iris.iloc[0, 4] = 0.2

#CONFERIR QUE O VALOR FOI ADICIONADO
iris.info()
print("\n")

#MOSTRAR QUANTAS LINHAS HÁ PARA CADA CATEGORIA DE ESPÉCIE - USADO PARA VERIFICAR BALANCEAMENTO
print(iris['Species'].value_counts())
print("\n")

#SUBSTITUIR OS VALORES DE UMA COLUNA POR OUTROS
iris['Species'] = iris['Species'].replace(
    {'Iris-setosa': 2, 'Iris-versicolor': 0, 'Iris-virginica': 1})

#MOSTRAR QUANTAS LINHAS HÁ PARA CADA CATEGORIA DE ESPÉCIE - AGORA COM OS VALORES SUBSTITUIDOS
print(iris['Species'].value_counts())
print("\n")

#IMPRIMIR O GRÁFICO SCATTERPLOT
sns.scatterplot(data=iris, x="SepalLengthCm", y="SepalWidthCm", hue="Species")

#GERAR UMA MATRIZ DE GRÁFICOS SCATTERPLOT
df = sns.load_dataset("iris")
sns.pairplot(df, hue="species")

#TREINAMENTO E TESTE
x = iris.iloc[:, 1:5]
y = iris['Species']

# LISTAS PARA MÉTRICAS
acc_list = []
prec_list = []
rec_list = []
f1_list = []

# HOLDOUT MANUAL
for i in range(5):
    x_train, x_test, y_train, y_test = train_test_split(
      x, y, test_size=0.30, random_state=i
    )

    #train_test_split(y, shuffle=False) #divide os dados no conjunto de treinamento e teste na ordem original, sem embaralhá-los. Diferente do padrão (shuffle=True).
    #Essencial quando a ordem cronológica importa, garantindo que o modelo treine com o passado e teste com o futuro.

    print(f"\n-------------- Iteração {i} --------------\n")

    # DECISION TREE --------------
    dt = DecisionTreeClassifier()

    dt.fit(x_train, y_train) #função que vai fazer o treinamento do modelo

    predict = dt.predict(x_test) #função que vai fazer o teste do modelo

    print("Predict (Decision Tree):\n", predict)
    print("\n")

    #MATRIZ DE CONFUSÃO
    conf_matrix = confusion_matrix(y_test, predict) #Retorna 3x3 pois há 3 classes

    disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=sorted(y.unique()))
    disp.plot(cmap=plt.cm.Blues)
    plt.title(f'Confusion Matrix - Decision Tree (Iteração {i})')
    plt.show()

    #ACCURACY
    accuracy = accuracy_score(y_test, predict)

    #PRECISION
    precision = precision_score(y_test, predict, average='macro') #dados balanceados: macro; dados desbalanceados: micro

    #RECALL
    recall = recall_score(y_test, predict, average='macro')

    #F1 SCORE
    f1 = f1_score(y_test, predict, average='macro')

    #ARMAZENAR MÉTRICAS
    acc_list.append(accuracy)
    prec_list.append(precision)
    rec_list.append(recall)
    f1_list.append(f1)

    #TABELA DE CLASSIFICAÇÃO, avaliar e comparar modelos.
    print("Decision Tree:\n", classification_report(y_test, predict))

    # SVM --------------
    svm = SVC()

    svm.fit(x_train, y_train) #função que vai fazer o treinamento do modelo

    predict = svm.predict(x_test) #função que vai fazer o teste do modelo

    print("Predict (SVM):\n", predict)
    print("\n")

    #MATRIZ DE CONFUSÃO
    conf_matrix = confusion_matrix(y_test, predict)

    disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=sorted(y.unique()))
    disp.plot(cmap=plt.cm.Blues)
    plt.title(f'Confusion Matrix - SVM (Iteração {i})')
    plt.show()

    #ACCURACY
    accuracy = accuracy_score(y_test, predict)

    #PRECISION
    precision = precision_score(y_test, predict, average='macro')

    #RECALL
    recall = recall_score(y_test, predict, average='macro')

    #F1 SCORE
    f1 = f1_score(y_test, predict, average='macro')

    #ARMAZENAR MÉTRICAS
    acc_list.append(accuracy)
    prec_list.append(precision)
    rec_list.append(recall)
    f1_list.append(f1)

    #TABELA DE CLASSIFICAÇÃO, avaliar e comparar modelos.
    print("SVM:\n", classification_report(y_test, predict))

    #LOOP MANUAL
    print(f"Iteration {i} (random_state={i}):")
    print(f"Test Indices: {x_test}")



print("\nMÉDIAS")
print(f"Accuracy: {np.mean(acc_list):.4f}")
print(f"Precision: {np.mean(prec_list):.4f}")
print(f"Recall: {np.mean(rec_list):.4f}")
print(f"F1 Score: {np.mean(f1_list):.4f}")

print("\nDESVIO PADRÃO")
print(f"Accuracy: {np.std(acc_list):.4f}")
print(f"Precision: {np.std(prec_list):.4f}")
print(f"Recall: {np.std(rec_list):.4f}")
print(f"F1 Score: {np.std(f1_list):.4f}")
```
